```python
##原版
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostRegressor, RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
from sklearn.linear_model import Ridge, Lasso, ElasticNet
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error, r2_score

random_state = 5064

# 切分訓練集和測試集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state = random_state)

#model = SVR(kernel='linear',C=0.35,epsilon=0.09)
#model = SVR(kernel='linear')
model = SVR(kernel='rbf',C=3,gamma=0.0002,epsilon=0.016,max_iter=2000)

# 特徵縮放
#計算均值和標準差：對於每個特徵，計算其所有樣本的均值和標準差。
#將數據進行轉換：對於每個特徵，將其數值減去均值，然後除以標準差。
#使得數據的每個特徵具有相同的尺度，從而避免了某些特徵尺度過大對模型造成的影響，也使得不同特徵之間的比較更加合理。
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 訓練模型
model.fit(X_train_scaled, y_train)

# 在訓練集和測試集上進行預測
y_train_pred = model.predict(X_train_scaled)
y_test_pred = model.predict(X_test_scaled)

# 計算MSE和R平方
mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)
r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)

print('SVR')
print(random_state)
print(f"train_r2:  {r2_train}")
print(f"test_r2:   {r2_test}")
print(f"train_mse: {mse_train}")
print(f"test_mse:  {mse_test}")
```
